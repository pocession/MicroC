```python
## Change your input file by changing the index
## Also the output file name is determined by the input file name

hicfile = data_dir + os.listdir(data_dir)[2]
hicfile

hic = hicstraw.HiCFile(hicfile)
```

```python
## Check information

print(hic.getGenomeID())
print(hic.getResolutions())
```

```python
## Check chromosomes

for chrom in hic.getChromosomes():
  print(chrom.name, chrom.length)
```

```python
## Grab the chr2 in 5kb solution
## With KR normalization
mzd = hic.getMatrixZoomData('chr2', 'chr2', "observed", "KR", "BP", 5000)
```

```python
## Get the IL1B super TAD
## This is to capture a snapshot from a interaction matrix
## .getRecordsAsMatrix(rowStart, rowEnd, columnStart, columnEnd)
numpy_matrix = mzd.getRecordsAsMatrix(112735986, 113204585, 112735986, 113204585)
```

```python
## output dir
output_dir = wd + "/../Results/processing/"
## Extract the last part after the last slash, then remove the extension
file_name_with_extension = hicfile.split('/')[-1]
outputfile = file_name_with_extension.rsplit('.', 1)[0]
outputfile = output_dir + outputfile + "_extracted.csv"

np.savetxt(outputfile, numpy_matrix, delimiter=",")
```

```python
## Generate the interaction starting position index
## And write into a csv file

start = 112735986
end = 113204585
resolution = 5000
fragment_num = math.ceil((end - start)/resolution)
fragment_index = []
fragment_index.append(start+1)
for i in range(fragment_num-2):
    new_start = fragment_index[i] + 5000
    fragment_index.append(new_start)
 
data = fragment_index
with open(output_dir + "start_position_index.txt", "w") as txt_file:
    for line in data:
        txt_file.write(str(line) + "\n")
```
